### Ejercicio 16: Crea una función para sumar dos tensores. Si las dimensiones son incompatibles debe devolver un mensaje de error.

```python
def sum_tensor_1(tensor_1, tensor_2):
  if tensor_1.shape != tensor_2.shape:
    raise ValueError('Los tensores tienen que tener las mismas dimensiones.')
  return tensor_1 + tensor_2

def sum_tensor_2(tensor_1, tensor_2):
  if tensor_1.shape != tensor_2.shape:
    raise ValueError('Los tensores tienen que tener las mismas dimensiones.')
  return torch.add(tensor_1, tensor_2)
```

### Ejercicio 17: Crea una función para restar dos tensores. Si las dimensiones son incompatibles debe devolver un mensaje de error.

```python
def subtract_tensors_1(tensor_1, tensor_2):
  if tensor_1.shape != tensor_2.shape:
    raise ValueError('Los tensores tienen que tener las mismas dimensiones.')
  return tensor_1 - tensor_2

def subtract_tensors_2(tensor_1, tensor_2):
  if tensor_1.shape != tensor_2.shape:
    raise ValueError('Los tensores tienen que tener las mismas dimensiones.')
  return torch.subtract(tensor_1, tensor_2)
```

*   Número de GPU (por si hay varias)
*   Nombre
*   Memoria Total
*   Capacidad de cómputo
*   Multiprocesador



```python
def print_gpu_properties(device):
    props = torch.cuda.get_device_properties(device)
    print(f"  Name: {props.name}")
    print(f"  Total Memory: {props.total_memory / (1024 ** 3):.2f} GB")
    print(f"  Compute Capability: {props.major}.{props.minor}")
    print(f"  Multiprocessors: {props.multi_processor_count}")
```

### Ejercicio 38: Crear una función que imprima la memoria de la GPU ocupada y la memoria reservada en GB

```python
def print_memory_use(device):
  print(f"Total Memory: {torch.cuda.get_device_properties(device).total_memory / (1024 ** 3):.2f} GB")
  print(f"Used memory: {torch.cuda.memory_allocated() / (1024 ** 3):.2f} GB")
  print(f"Reserved memory: {torch.cuda.memory_reserved() / (1024 ** 3):.2f} GB")
```

```python
print_memory_use(0)
```

```python
class LinearLayer(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearLayer, self).__init__()

        self.weight = Parameter(torch.randn(output_dim, input_dim))
        self.bias = Parameter(torch.randn(output_dim))

    def forward(self, x):

        return x @ self.weight.T + self.bias
```
